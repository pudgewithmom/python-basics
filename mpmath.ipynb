# mpmath
Multiple Precision math

Пакет для работы с числами с плавающей точкой со сколь угодно высокой точностью. В нём реализованы алгоритмы вычисления элементарных функций, а также большого количества специальных функций.

```python
from mpmath import *
%matplotlib inline
```

Точность контролируется глобальным объектом `mp`.

```python
print(mp)
```

`prec` - число бит в мантиссе, `dps` - число значащих десятичных цифр. Если изменить один из этих атрибутов, другой изменится соответственно.

```python
mp.dps=50
print(mp)
```

`mpf` создаёт число с плавающей (multiple precision float) точкой из строки или числа.

```python
x=mpf('0.1')
print(x)
```

```python
x
```

А вот так делать не надо. `0.1` сначала преобразуется в число с плавающей точкой со стандартной (т.е. двойной) точностью, а потом уже оно преобразуется в `mpf`.

```python
y=mpf(0.1)
print(y)
```

Чтобы не потерять точность, нужно делать `mpf` из строки или из отношения целых чисел (вероятно, со знаменателем вида $10^n$).

```python
y=mpf(1)/10
print(y)
```

## Интегралы

При вычислении этого интеграла все вычисления будут производиться с точностью, на 5 значащих цифр большей; затем она вернётся к прежней.

```python
with extradps(5):
    I=quad(lambda x:log(x)**2/(1+x),(0,1))
    print(I)
```

Допустим, у нас есть причины подозревать, что этот интеграл равен $\zeta(3)$, умноженному на рациональное число (с не очень большими числителем и знаменателем). `pslq([x1,x2,...])` находит целые числа $n_1$, $n_2$, ... такие, что $n_1\,x_1 + n_2\,x_2 + \cdots = 0$. Это - метод нахождения тождеств, называемый *экспериментальной математикой*. Для этого часто требуются вычисления с очень высокой точностью.

```python
pslq([I,zeta(3)])
```

То есть наш интеграл равен $\frac{3}{2} \zeta(3)$. Это, конечно, не доказательство. Но если мы ещё увеличим точность вычисления интеграла, а результат `pslq` не изменится, то мы можем быть практически уверены, что этот результат верен.

Двойной интеграл:

```python
quad(lambda x,y:1/(1+x*y),[0,1],[0,1])
```

## Сумма ряда

```python
with extradps(5):
    s=nsum(lambda n:(-1)**(n-1)/n**4,(1,inf))
    print(s)
```

```python
pslq([s,pi**4])
```

То есть эта сумма, вероятно, равна $\frac{7}{720} \pi^4$.

Точки, в которых функция вызывает числовое исключение или возвращает бесконечное значение, удаляются с графика. Сингулярности также можно явно исключить следующим образом (полезно для удаления ошибочных вертикальных линий):

```python
plot(cot, ylim=[-5, 5])   # bad
plot(cot, ylim=[-5, 5], singularities=[-pi, 0, pi])  # good
```

Для частей, где функция принимает комплексные значения, действительная часть изображается штрихами, а мнимая часть — точками.

### Графики сложных функций

```python
fp.cplot(fp.gamma, points=100000)
```

`mpmath.cplot( f , re = [- 5, 5] , im = [- 5, 5] , points = 2000 , color = None , verbose = False , file = None , dpi = None , axes = None )`

Наносит заданную комплексную функцию f на прямоугольную часть комплексной плоскости, заданную парами интервалов re и im . Например:

```python
cplot(lambda z: z, [-2, 2], [-10, 10])
cplot(exp)
cplot(zeta, [0, 1], [0, 50])
```

По умолчанию сложный аргумент (фаза) отображается как цвет (оттенок), а величина отображается как яркость. Вы также можете указать пользовательскую функцию цвета ( color ). Эта функция должна принимать в качестве входных данных комплексное число и возвращать 3-кортеж RGB, содержащий числа с плавающей запятой в диапазоне 0,0–1,0.

Кроме того, вы можете выбрать встроенную функцию цвета, передав строку в качестве цвета :

* «default» - цветовая схема по умолчанию

* «фаза» — цветовая схема, отображающая только фазу функции,
с белым для положительных реалов, черным для отрицательных реалов, золотым в верхней полуплоскости и синим в нижней полуплоскости.

Для получения четкого изображения может потребоваться увеличить количество точек до 100 000 или около того. Поскольку многократная оценка функции, вероятно, будет медленной, параметр «подробный» полезен для отображения хода выполнения.
